```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good devoloper',
    'Understand the simple rnn',
    'you are a nice media editor'
]
```

```python
### Defining the vocabulary size
voc_size = 10000
```

```python
one_hot_repr = [one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
### Word Embedding Representation

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

We have to make all the sentences to the equal size while training with rnn  
at last all the words that are going it will be fixed number of time stamps based on the sentence size.so,pad_sequences is useful for this problem

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

Parameter -> padding='pre'  
***Definition:*** Specifies where padding should be applied.  
- 'pre' means that zeros will be added before each sequence (at the beginning), effectively shifting the sequence to the right.  
- 'post' would mean adding zeros after each sequence (at the end).

```python
### Feature Representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
